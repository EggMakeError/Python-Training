```python
import pandas as pd
```

Read the csv file from this url:
'https://drive.google.com/file/d/1P2EyVCFAPeJcQ_lAN2W-bjwtbizOeFXE/view?usp=drive_link'

Name it `sets_df`

```python
url ='https://drive.google.com/file/d/1P2EyVCFAPeJcQ_lAN2W-bjwtbizOeFXE/view?usp=drive_link' # google drive link of the uploaded file goes here
path = 'https://drive.google.com/uc?export=download&id=' +url.split('/')[-2] # the command after the download command is the code for selecting the data after the slashes and the -2 selects the second to last data set after the last slash
sets_df = pd.read_csv(path, delimiter=',')  # you have to actually define the delimiter because otherwise it will throw a parsing error. Normally, defining the delimiter as "';'" is enough but in this csv, the delimiter to use is "','" because its using "," to separate the data sets for some reason.

sets_df
```

#### Exercise 1
Select and display only the `"name"` and `"year"` columns from `sets_df` using `.loc`

```python
result = sets_df.loc[:, ["name", "year"]]

print(result.to_string(index=False))
```
