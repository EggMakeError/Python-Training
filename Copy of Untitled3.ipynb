# Data Cleaning Section

**orderlines_df cleaning**

```python
# we will start by dropping columns that contain no data

orderlines_df.drop(columns=['product_id'], inplace=True)

# next I want to rename the id_order column from orders_df to order_id so it matches up with orderlines_df

orderlines_df = orderlines_df.rename(columns={'id_order': 'order_id'})

# lets do the same for the 'date' column in orderlines_df

orderlines_df['date'] = pd.to_datetime(orderlines_df['date'])

# in order to convert the object type of the unit_price column in orderlines_df to numeric, we have to first drop the additional decimal points in line 6 and beyond

orderlines_df['unit_price'] = orderlines_df['unit_price'].str.replace(r'(?<=\d)\.(?=.*\.)', '', regex=True)

# Convert to numeric so we can work with it later on

orderlines_df['unit_price'] = pd.to_numeric(orderlines_df['unit_price'], errors='coerce')
```

**orders_df cleaning**

```python
# we want to change the "created_date" column in orders_df to the datetime format

orders_df['created_date'] = pd.to_datetime(orders_df['created_date'])

# in orders_df there are 5 lines missing a value for "total_paid", we are going to locate those 5 to assign a value

specific_order_ids = [427314, 431655, 447411, 448966, 449596]
filtered_orders = orderlines_df.loc[orderlines_df['order_id'].isin(specific_order_ids)]

print(filtered_orders)

missing_total_paid = orders_df.loc[orders_df['total_paid'].isna()]

# first, lets take a look at these lines to better understand the structure and also to see the common denominator which we can use to find the actual item value
print(missing_total_paid)

# we want to reassign the values from the unit_price column from orderlines_df to the missing total_paid in orders_df

for order_id in specific_order_ids:

    unit_price_reassign = orderlines_df.loc[orderlines_df['order_id'] == order_id, 'unit_price'].astype(float).sum()
    orders_df.loc[orders_df['order_id'] == order_id, 'total_paid'] = unit_price_reassign

print(orders_df.loc[orders_df['order_id'].isin(specific_order_ids)])
```

**products_df cleaning**

```python
# here we are doin the same thing we did in the orderlines_df, just dropping the second decimal points in products_df price and promo_price which we have to do by converting them to a string first and then reverting it back to a float64 nvalue

products_df['price'] = products_df['price'].replace(r'(?<=\d)\.(?=.*\.)', '', regex=True).astype(str)
products_df['promo_price'] = products_df['promo_price'].replace(r'(?<=\d)\.(?=.*\.)', '', regex=True).astype(str)

products_df['price'] = pd.to_numeric(products_df['price'], errors='coerce')
products_df['promo_price'] = pd.to_numeric(products_df['promo_price'], errors='coerce')
```

```python
# we are creating a filter to apply to the dataframe filtering all the promo_price values that are higher than the initial price (because a promotion price should be lower than the initial price)

mask_promo_higher =  pd.to_numeric(products_df['promo_price'], errors='coerce') > pd.to_numeric(products_df['price'], errors='coerce')

# we are locating all the lines that adhere to our filter without touching the values that are already lower than the initial price (using pd.to_numeric to only change the "object" type of the filtered values to float64 ensuring that the values that are correct, don't get touched)

products_df.loc[mask_promo_higher, 'promo_price'] = pd.to_numeric(products_df.loc[mask_promo_higher, 'promo_price'], errors='coerce')

# we are now applying a calculation to these located lines because what I think happened is that instead of a discount, they applied the discount value as a multiplier (i.e. instead of multiplying with 0.82, they multiplied with 8.2 because the decimal point was not at the correct location)

products_df.loc[mask_promo_higher, 'promo_price'] = (products_df.loc[mask_promo_higher, 'promo_price'] / 10).round(2)

# and now we convert all the values in 'promo_price' to numeric (in this case float64)
products_df['promo_price'] = pd.to_numeric(products_df['promo_price'], errors='coerce')

# rounding the last 2 digits in each line so it shows a decimal point
products_df['promo_price'] = products_df['promo_price'].round(2)
```
