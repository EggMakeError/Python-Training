# **Pythomon Game**

This is a coding exercise where the goal is to create a game akin to a creature catching game al√° Pokemon (without using any assets or any of the copyrighted game mechanics of course) in which you catch creatures by solving faulty python code.

The trick here is that you can carry a team of up to 6 of these creatures where every creature can hold up to 4 attacks which are basically the python command lines you are allowed to use to solve the faulty code.

The game would work in a way that when you encounter a creature, these creatures pose a coding challenge to you (the player) with a code segment and multiple (1-20) missing python command lines. Every round you can fill as many of the spots as you want with "attacks" from the attack list of your captured creatures and you do damage according to how many spots were filled with the correct command line. The opposing creature then also does damage to your creature based on how many errors you "missed" or you have put in incorrectly.

**Step 1: Planning and Conceptualizing the Game Design**

Before jumping into code, we need to plan the core components and mechanics of the game. These include:

    Player and Creature Mechanics:
        Each player has a team of up to 6 creatures.
        Each creature has 4 attacks (command lines).
        Each attack corresponds to Python commands (e.g., print(), len(), append()).

    Battle System:
        An opponent creature presents a coding challenge.
        The challenge consists of faulty code with missing or incorrect lines.
        The player uses "attacks" (Python commands) to fill the gaps.
        Correct commands deal damage; missed or incorrect lines result in damage taken.

    Game Flow:
        Players encounter creatures randomly.
        Solving challenges rewards the player with new creatures or attacks.
        The game ends when the player's creatures are defeated or they "catch 'em all."

    Data Storage:
        Challenges (faulty code) and solutions need to be stored in a structured way.
        Creatures, their attacks, and stats (HP, level, damage) must also be tracked.
