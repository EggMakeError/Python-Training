# Data Cleaning Section

**brands_df cleaning**

```python
# In brands_df, there is actually not a whole lot to do, all we have to do is find out which long descriptors got either mapped to the wrong short descriptors or contain "unknown" for a value

# Group by the 'long' column and count unique 'short' values
non_unique_long = (
    brands_df.groupby('long')['short']
    .nunique()
    .reset_index(name='unique_count')
)

# Filter for rows where the count of unique 'short' values is greater than 1
non_unique_long = non_unique_long[non_unique_long['unique_count'] > 1]

# Get the rows from the original dataframe that match these 'long' values
non_unique_rows = brands_df[brands_df['long'].isin(non_unique_long['long'])]

print(non_unique_rows)

# now that we know which values are affected, we can then find out if the short descriptors have a corresponding part in the products_df desc column

# products_df[products_df['sku'].str[:3] == 'MUJ']

# after we found out which companies or products the descriptors actually correspond to, we can then map new values to the short column

brands_mapping = {
    "Apple Second Hand": "AP2",
    "Computer Aided Design": "CAD",
    "Environmental": "ENV",
    "Kensington": "KEN",
    "Others": "OTR",
    "Library": "LIB",
    "Mujjo": "MUJ"

}

# Replace the long names with the names from the mapping library
brands_df["long"] = brands_df["short"].map(brands_mapping).fillna(brands_df["short"])
```

**orderlines_df cleaning**

```python
# we will start by dropping columns that contain no data

orderlines_df.drop(columns=['product_id'], inplace=True)

# next I want to rename the id_order column from orders_df to order_id so it matches up with orderlines_df

orderlines_df = orderlines_df.rename(columns={'id_order': 'order_id'})

# lets do the same for the 'date' column in orderlines_df

orderlines_df['date'] = pd.to_datetime(orderlines_df['date'])

# in order to convert the object type of the unit_price column in orderlines_df to numeric, we have to first drop the additional decimal points in line 6 and beyond

orderlines_df['unit_price'] = orderlines_df['unit_price'].str.replace(r'(?<=\d)\.(?=\d+\.\d)', '', regex=True)

# we are converting lowercase letters to upper case in the SKU to prevent mismatches and for readability
orderlines_df['sku'] = orderlines_df['sku'].str.strip().str.upper()

# TODO: include explanation for regex argument

# Convert to numeric so we can work with it later on

orderlines_df['unit_price'] = pd.to_numeric(orderlines_df['unit_price'], errors='coerce')
```

**orders_df cleaning**

```python
# we want to change the "created_date" column in orders_df to the datetime format

orders_df['created_date'] = pd.to_datetime(orders_df['created_date'])

# in orders_df there are 5 lines missing a value for "total_paid", we are going to locate those 5 to assign a value

missing_total_paid = orders_df.loc[orders_df['total_paid'].isna()]

print(missing_total_paid)

# first, lets take a look at these lines to better understand the structure and also to see the common denominator which we can use to find the actual item value

specific_order_ids = [427314, 431655, 447411, 448966, 449596]
filtered_orders = orderlines_df.loc[orderlines_df['order_id'].isin(specific_order_ids)]

print(filtered_orders)

# we want to reassign the values from the unit_price column from orderlines_df to the missing total_paid in orders_df

for order_id in specific_order_ids:

    unit_price_reassign = orderlines_df.loc[orderlines_df['order_id'] == order_id, 'unit_price'].astype(float).sum()
    orders_df.loc[orders_df['order_id'] == order_id, 'total_paid'] = unit_price_reassign

print(orders_df.loc[orders_df['order_id'].isin(specific_order_ids)])
```

**products_df cleaning**

```python
# first, lets drop all the duplicated rows of which there are over 8000

products_df.drop_duplicates(subset='sku', keep='first', inplace=True)
```

```python
# here we are doing the same thing we did in the orderlines_df, just dropping the second decimal points in products_df price and promo_price which we have to do by converting them to a string first and then reverting it back to a float64 nvalue

products_df['price'] = products_df['price'].replace(r'(?<=\d)\.(?=.*\.)', '', regex=True).astype(str)
products_df['promo_price'] = products_df['promo_price'].replace(r'(?<=\d)\.(?=.*\.)', '', regex=True).astype(str)

products_df['price'] = pd.to_numeric(products_df['price'], errors='coerce')
products_df['promo_price'] = pd.to_numeric(products_df['promo_price'], errors='coerce')

# we are converting lowercase letters to upper case in the SKU to prevent mismatches and for readability
products_df['sku'] = products_df['sku'].str.strip().str.upper()
```

```python
# we are creating a filter to apply to the dataframe filtering all the promo_price values that are higher than the initial price (because a promotion price should be lower than the initial price)

mask_promo_higher =  pd.to_numeric(products_df['promo_price'], errors='coerce') > pd.to_numeric(products_df['price'], errors='coerce')

# we are locating all the lines that adhere to our filter without touching the values that are already lower than the initial price (using pd.to_numeric to only change the "object" type of the filtered values to float64 ensuring that the values that are correct, don't get touched)

products_df.loc[mask_promo_higher, 'promo_price'] = pd.to_numeric(products_df.loc[mask_promo_higher, 'promo_price'], errors='coerce')

# we are now applying a calculation to these located lines because what I think happened is that instead of a discount, they applied the discount value as a multiplier (i.e. instead of multiplying with 0.82, they multiplied with 8.2 because the decimal point was not at the correct location)

products_df.loc[mask_promo_higher, 'promo_price'] = (products_df.loc[mask_promo_higher, 'promo_price'] / 10).round(2)

# and now we convert all the values in 'promo_price' to numeric (in this case float64)
products_df['promo_price'] = pd.to_numeric(products_df['promo_price'], errors='coerce')

# rounding the last 2 digits in each line so it shows a decimal point
products_df['promo_price'] = products_df['promo_price'].round(2)

# we will fill the NaN values in price

missing_values_price = products_df.loc[products_df['price'].isna(), 'sku']

for sku in missing_values_price:

  product_price_reassign = orderlines_df.loc[orderlines_df['sku'] == sku, 'unit_price']

if not product_price_reassign.empty:
  products_df.loc[products_df['sku'] == sku, 'price'] = product_price_reassign



print(missing_values_price)



# Since some prices in the price column are heavily inflated, we want to Calculate the ratio between price and promo_price and then reapply the newly calculated values back to the price column
def adjust_price(row):
    ratio = row['price'] / row['promo_price']

    if ratio >= 9.5:
        return round(row['price'] / ratio, 2)
    else:
        return row['price']

products_df['price'] = products_df.apply(adjust_price, axis=1)

adjusted_rows = products_df[products_df['price'] != products_df['promo_price']]

print(adjusted_rows)

# We can decide to fill the remaining 22 missing prices with the promo price values

products_df['price'] = products_df['price'].fillna(products_df['promo_price'])

# some of the descriptions are missing from the description column. We will fill them with the values from "name" since the names are already very descriptive and vice versa
products_df['desc'] = products_df['desc'].fillna(products_df['name'])
products_df['name'] = products_df['name'].fillna(products_df['desc'])

# Identify SKUs that are in orderlines_df but not in products_df
missing_skus = orderlines_df[~orderlines_df['sku'].isin(products_df['sku'])]

# For each missing SKU, calculate the maximum unit_price, mean unit_price, and create the new row
new_rows = missing_skus.groupby('sku').agg(
    price=('unit_price', 'max'),
    promo_price=('unit_price', 'mean')
).reset_index()

# Add the columns 'name', 'desc', and 'in_stock' to the new rows
new_rows['name'] = 'None'
new_rows['desc'] = 'None'
new_rows['in_stock'] = 0
new_rows['type'] = 'None'  # Add any additional column if needed

products_df = pd.concat([products_df, new_rows], ignore_index=True)

products_df['price'] = products_df['price'].round(2)
products_df['promo_price'] = products_df['promo_price'].round(2)
```

```python
import pandas as pd
from collections import Counter
import re

# Define the function to assign categories
def assign_category_with_checks(df):
    # Keywords and their corresponding categories
    keyword_categories = {
        'laptop': 'Laptops',
        'phone': 'Phones',
        'tablet': 'Tablets',
        'headphone': 'Headphones',
        'charger': 'Chargers',
        'case': 'Cases',
        'ram': 'RAM Modules',
        'ssd': 'SSDs',
        'keyboard': 'Keyboards',
        'monitor': 'Monitors'
    }

    # Regular expressions for filtering false positives
    false_positive_patterns = {
        'Laptops': r'notebook|case|bag',
        'Phones': r'sim|accessory|cover',
        'RAM Modules': r'software|system',
        # Add patterns for other categories if needed
    }

    # Store category and type mappings
    category_types = {}

    # Combine keywords to detect and assign categories
    def detect_category(row):
        combined_text = f"{row['name']} {row['desc']}".lower()

        # Match category by keyword
        for keyword, category in keyword_categories.items():
            if keyword in combined_text:
                return category
        return 'Other'

    # Apply initial category detection
    df['category'] = df.apply(detect_category, axis=1)

    # Identify the most common type for each category
    for category in df['category'].unique():
        if category != 'Other':
            most_common_type = (
                df[df['category'] == category]['type']
                .mode()
                .iloc[0]
            )
            category_types[category] = most_common_type

    # Validate entries based on type and false-positive patterns
    def validate_category(row):
        category = row['category']
        if category == 'Other':
            return 'Other'

        # Check if type matches the most common type
        if row['type'] != category_types.get(category):
            return 'Other'

        # Check for false positives using regex
        pattern = false_positive_patterns.get(category)
        if pattern and re.search(pattern, f"{row['name']} {row['desc']}".lower()):
            return 'Other'

        return category

    # Apply validation to refine the categories
    df['category'] = df.apply(validate_category, axis=1)

    return df

# Use the function on the products DataFrame
products_df = assign_category_with_checks(products_df)

# Output the updated DataFrame
print(products_df[['name', 'desc', 'type', 'category']].head(20))
```

```python
import pandas as pd
import re
from collections import Counter

def assign_better_category(df):
    # Expanded keyword mappings
    keyword_categories = {
        'laptop|notebook': 'Laptops',
        'phone|iphone|android|smartphone': 'Phones',
        'tablet|ipad': 'Tablets',
        'headphone|earbud|audio': 'Headphones',
        'charger|adapter|magsafe|power supply': 'Chargers',
        'case|cover|shell': 'Cases',
        'ram|memory|ddr[1-4]': 'RAM Modules',
        'ssd|solid state drive': 'SSDs',
        'keyboard|keypad': 'Keyboards',
        'monitor|screen|display': 'Monitors',
        # Additional categories
        'dock|docking station': 'Docks',
        'cable|connector|usb': 'Cables',
    }

    # False positive patterns
    false_positive_patterns = {
        'Laptops': r'bag|case|cover|stand',
        'Phones': r'accessory|dock|cable|case',
        'RAM Modules': r'software|non-ram',
        'Chargers': r'adapter.*not charger|battery',
        'Cables': r'accessory|case',
    }

    # Detect category by keywords
    def detect_category(row):
        combined_text = f"{row['name']} {row['desc']}".lower()
        for pattern, category in keyword_categories.items():
            if re.search(pattern, combined_text):
                return category
        return 'Other'

    # Apply category detection
    df['category'] = df.apply(detect_category, axis=1)

    # Determine the most common type for each category
    category_types = {}
    for category in df['category'].unique():
        category_df = df[df['category'] == category]
        most_common_type = category_df['type'].mode()
        if not most_common_type.empty:
            category_types[category] = most_common_type.iloc[0]

    # Validate and refine categories
    def validate_category(row):
        category = row['category']
        combined_text = f"{row['name']} {row['desc']}".lower()

        # Check if type matches the most common type
        if category in category_types and row['type'] != category_types[category]:
            return 'Other'

        # Check for false positives
        pattern = false_positive_patterns.get(category)
        if pattern and re.search(pattern, combined_text):
            return 'Other'

        return category

    # Refine categories with validation
    df['category'] = df.apply(validate_category, axis=1)

    return df

# Apply the refined function
products_df = assign_better_category(products_df)

# Output the refined DataFrame
print(products_df[['name', 'desc', 'type', 'category']].head(20))
```

```python
import pandas as pd
from collections import Counter
import re
from nltk.corpus import stopwords

# Download stopwords (if not already downloaded)
import nltk
nltk.download('stopwords')

# Get the set of English stopwords
stop_words = set(stopwords.words('english'))

# Function to clean text, tokenize, and remove stopwords
def clean_and_tokenize(text):
    # Lowercase, remove non-alphanumeric characters, and split into words
    words = re.findall(r'\b\w+\b', text.lower())
    # Remove stopwords from the words list
    filtered_words = [word for word in words if word not in stop_words]
    return filtered_words

# Function to get most common words for each type
def get_common_words_by_type(df, top_n=5):
    # Dictionary to hold common words for each type
    common_words_by_type = {}

    # Group by 'type'
    grouped_by_type = df.groupby('type')

    for type_id, group in grouped_by_type:
        # Combine 'name' and 'desc' columns into one text column
        combined_texts = group['name'] + ' ' + group['desc']

        # Tokenize and clean all the combined texts for this type
        all_words = []
        for text in combined_texts:
            words = clean_and_tokenize(text)
            all_words.extend(words)

        # Count word frequencies for this type
        word_counts = Counter(all_words)

        # Get the most common words for this type
        common_words = word_counts.most_common(top_n)  # Top N most common words
        common_words_by_type[type_id] = [word for word, _ in common_words]

    return common_words_by_type

# Function to categorize each row based on the most common words of each type
def categorize_based_on_keywords(df, common_words_by_type):
    def assign_category(row):
        # Combine 'name' and 'desc' columns into one text
        text = row['name'] + ' ' + row['desc']
        words = clean_and_tokenize(text)

        # Check which category matches based on common words
        for type_id, common_words in common_words_by_type.items():
            # If any word in the text matches a common word for a type, assign that type
            if any(word in common_words for word in words):
                return type_id
        return 'Uncategorized'  # Default if no match is found

    # Apply the categorization function to each row and create a new 'category' column
    df['category'] = df.apply(assign_category, axis=1)

    return df

# Sample dataframe (assuming the dataframe looks something like this)

# Get common words for each type (with stopwords removed)
common_words_by_type = get_common_words_by_type(products_df, top_n=5)

# Display the common words for each type
print("Most common words for each type:")
for type_id, words in common_words_by_type.items():
    print(f"Type: {type_id}")
    print(f"  Most common words: {words}")
    print("\n")

# Categorize products based on the most common words (with stopwords removed)
categorized_df = categorize_based_on_keywords(products_df, common_words_by_type)

# Display the updated dataframe with categories
print("Updated DataFrame with Categories:")
print(categorized_df)
```

```python
predefined_categories = {
    "Laptops": {
        "include": ["macbook", "imac", "ssd", "core", "pro"],
        "exclude": ["iphone", "ipad", "case", "repair"]
    },
    "Phones": {
        "include": ["iphone", "android", "repair", "service"],
        "exclude": ["macbook", "imac", "ssd"]
    },
    "Tablets": {
        "include": ["ipad", "mini", "pro", "wi", "fi"],
        "exclude": ["iphone", "macbook", "case"]
    },
    "Accessories iPhone": {
        "include": ["case", "cable", "charger", "adapter", "screen"],
        "exclude": ["macbook", "imac", "iphone", "ipad"]
    },
    "Storage": {
        "include": ["hard", "drive", "usb", "ssd", "nas"],
        "exclude": ["iphone", "ipad", "case"]
    },
    "Accessories Mac": {
        "include": ["mouse", "cable", "apple"],
        "exclude": ["iphone", "ipad"]
    }
}

# Function to determine category based on common words
def assign_category_to_type_with_filters(common_words, predefined_categories):
    # Initialize variables to track the best category
    max_overlap = 0
    best_category = "Uncategorized"

    # Check each predefined category
    for category, filters in predefined_categories.items():
        include_words = set(filters["include"])
        exclude_words = set(filters["exclude"])

        # Calculate overlap with include words
        overlap = len(set(common_words) & include_words)

        # Check if any exclude words are present
        # Ensure at least 2 words match for assignment
        if overlap >= 2 and not (set(common_words) & exclude_words):
            if overlap > max_overlap:
                max_overlap = overlap
                best_category = category

    return best_category

# Create a mapping of type to category
type_to_category = {
    type_id: assign_category_to_type_with_filters(common_words, predefined_categories)
    for type_id, common_words in common_words_by_type.items()
}

# Assign categories to each row in products_df
products_df['category'] = products_df['type'].map(type_to_category)

# Display the updated DataFrame
products_df.head(50)
```
