```python
import pandas as pd
```

Read the csv file from this url:
'https://drive.google.com/file/d/1sXYEFFcJj0qlfqp08lPWaeuu_z35ICfp/view?usp=drive_link'

Name it `sets_df`

```python
url ='https://drive.google.com/file/d/1sXYEFFcJj0qlfqp08lPWaeuu_z35ICfp/view?usp=drive_link' # google drive link of the uploaded file goes here
path = 'https://drive.google.com/uc?export=download&id=' +url.split('/')[-2] # the command after the download command is the code for selecting the data after the slashes and the -2 selects the second to last data set after the last slash
sets_df = pd.read_csv(path, delimiter=',')  # you have to actually define the delimiter because otherwise it will throw a parsing error. Normally, defining the delimiter as "';'" is enough but in this csv, the delimiter to use is "','" because its using "," to separate the data sets for some reason.

sets_df
```

> Make sure you save the changes by overwriting the original DataFrame:  
e.g. `df['name'] = df['name'].str.upper()`

```python
sets_df['name'] = sets_df['name'].str.upper()
```

#### Exercise 6
There are strange symbols around the set numbers. Remove them.

```python
sets_df['set_num'] = sets_df['set_num'].str.strip() # removes the blank spaces

sets_df['set_num'] = sets_df['set_num'].fillna('').str.replace('[^a-zA-Z0-9-]', '', regex = True) # regex is used for this line of code but the default regex is changed from True to false in newer panda versions so you have to set it manually

sets_df['set_num']
```

#### Exercise 7
Set numbers are in two pieces, separated by '-'. Separate the two pieces from each other. Save the first part as a new column `set_num_prefix` and the second part as `set_num`.

```python
sets_df['set_num_prefix'] = sets_df['set_num'].str.split('-').str[0]
sets_df['set_num'] = sets_df['set_num'].str.split('-').str[1]

sets_df
```

#### Exercise 8
Replace the word 'brick' in the column `name` with the word 'block'.     
Check after to make sure the word 'brick' is no longer in any name.

```python
sets_df['name'] = sets_df['name'].str.replace('brick', 'block')

sets_df.loc[sets_df['name'].str.contains('brick')]
```
