```python
import pandas as pd

# Step 1: Create a population (list of specific numbers)
population = [5, 3, 6, 78, 1, 2, 57, 8, 231, 9, 45, 67, 89, 23, 56, 90, 12, 34, 76, 54, 32, 11, 19, 27, 38, 49, 59, 69, 79, 99]

# Step 2: Draw samples and compute sample means
n_samples = 10000  # Number of samples to draw
sample_size = 30   # Each sample consists of 30 values

sample_means = pd.Series([
    pd.Series(pd.Series(population).sample(n=sample_size, replace=True)).mean()
    for _ in range(n_samples)
])

# Step 3: Plot histogram of sample means
sample_means.plot(kind='hist', bins=50, density=True, alpha=0.6, color='g')
```
