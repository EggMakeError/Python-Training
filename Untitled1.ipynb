```python
import random
target_number = random.randint(1, 50)

guess = int(input("Guess the number (between 1 and 50): "))  # Add newline after this line

while guess != target_number:
    if guess < target_number:
        print("Too low!")
    else:
        print("Too high!")
    guess = int(input("Guess the number (between 1 and 50): "))

print("Correct!")
```
